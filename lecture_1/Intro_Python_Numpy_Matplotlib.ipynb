```
class BaseClass:
  def __init__(self,param1,param2):
    self.param1 = param1
    self.param2 = param2
    
  def get_param1(self):
    return self.param1
  
  def get_param2(self):
    return self.param2
    
```

```
class SubClass(BaseClass):
  def __init__(self,param1,param2,param3):
    super().__init__(param1,param2)
    self.param3 = param3
    
  def get_param3(self):
    return self.param3
```

```
sub = SubClass(1,2,3)
print(sub.get_param1(), sub.get_param2(), sub.get_param3())
```

#Hello Matplotlib

```
import matplotlib.pyplot as plt
```

```
#Generate 100 fake data points
def lin(w,b,x): 
  return w*x+b

def gen_fake_data(n, w, b):
    x = np.random.uniform(0,1,n) 
    y = lin(w,b,x) + 0.1 * np.random.normal(0,5,n)
    return x, y

x, y = gen_fake_data(100, 3., 8.)
```

```
plt.scatter(x,y,s=6); 
plt.xlabel("x"); 
plt.ylabel("y");
```
